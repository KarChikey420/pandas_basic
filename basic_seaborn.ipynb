```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# line plot with seaborn

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperature = [15, 17, 21, 26, 30, 32, 33, 31, 28, 24, 19, 16]
temp_df = pd.DataFrame({"Month": months, "Temperature": temperature})

plt.figure(figsize=(5,5))
sns.lineplot(x='Month', y='Temperature',data=temp_df,marker='o',color='b',linestyle='-')
plt.title("Average Monthly Temperature")
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.show()
```

# Bar Plot using seaborn

```python
roles = ["Data Engineer", "Data Analyst", "ML Engineer", "Backend Developer", "Frontend Developer"]
salary = [8.5, 6.2, 10.0, 9.0, 7.5]
salary_df = pd.DataFrame({"Role": roles, "Salary": salary})

plt.figure(figsize=(5,5))
sns.barplot(x='Role',y='Salary',data=salary_df,palette='viridis')
plt.title("Average Salary by Role(in Lakhs)")
plt.xlabel('Role')
plt.ylabel('Salary (in Lakhs)')
plt.xticks(rotation=15)
plt.show()
```

# Scatter plot using seaborn

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [45, 50, 55, 65, 70, 80, 85, 90]
study_df = pd.DataFrame({"Study Hours": study_hours, "Exam Score": exam_scores})
plt.Figure(figsize=(5,5))
sns.scatterplot(x='Study Hours', y='Exam Score',data=study_df,color='g',s=100)
sns.regplot(x='Study Hours',y='Exam Score',data=study_df,scatter=False,color='b')
plt.title("Study Hours vs Exam Score")
plt.show()
```

# Histogram chart using seaborn

```python
marks = np.random.normal(70, 10, 100) 
marks_df = pd.DataFrame({"Marks": marks})

sns.histplot(marks_df["Marks"],bins=10,kde=True,color="skyblue")
plt.title("Distrubution Of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```
